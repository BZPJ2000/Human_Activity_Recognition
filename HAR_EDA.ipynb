```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def perform_tsne(X_data, y_data, perplexities, n_iter=1000, img_name_prefix='t-sne'): 
    unique_labels = y_data.unique() 
    markers = ['^', 'v', 's', 'o', '1']  # 确保这里的标记数与 unique_labels 的数量相同 
    
    for index, perplexity in enumerate(perplexities):
        print('\nperforming tsne with perplexity {} and with {} iterations at max'.format(perplexity, n_iter)) 
        X_reduced = TSNE(verbose=2, perplexity=perplexity).fit_transform(X_data)
        print('Done..') 
        
        # prepare the data for seaborn         
        print('Creating plot for this t-sne visualization..')
        df = pd.DataFrame({'x': X_reduced[:, 0], 'y': X_reduced[:, 1], 'label': y_data})
        
        # draw the plot in appropriate place in the grid
        sns.lmplot(data=df,  x='x', y='y', hue='label', fit_reg=False, height=8, aspect=1, 
                   palette="Set1", markers=markers)  # 使用调整后的 markers
        plt.title("perplexity  : {} and max_iter : {}".format(perplexity, n_iter)) 
        img_name = img_name_prefix + '_perp_{}_iter_{}.png'.format(perplexity, n_iter) 
        print('saving this plot as image in present working directory...')
        plt.savefig(img_name)  
        plt.show() 
        print('Done')
 
# 假设 train 已经定义好，并且包含了 'subject', 'Activity', 'ActivityName' 等列 
X_pre_tsne = train.drop(['subject',  'Activity', 'ActivityName'], axis=1) 
y_pre_tsne = train['ActivityName'] 
perform_tsne(X_data=X_pre_tsne, y_data=y_pre_tsne, perplexities=[2, 5, 10, 20, 50])
```

```python
import seaborn as sns 
import matplotlib.pyplot  as plt 
from sklearn.manifold  import TSNE
import pandas as pd 
import os  # 用于处理文件路径 
 
def perform_tsne(X_data, y_data, perplexities=[2, 5, 10, 20, 50], n_iter=1000, img_name_prefix='tsne'): 
    # 确保输出目录存在 
    output_dir = 'tsne_outputs'
    if not os.path.exists(output_dir): 
        os.makedirs(output_dir) 
    
    for perplexity in perplexities: 
        try:
            tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=n_iter, random_state=42)
            X_reduced = tsne.fit_transform(X_data)  
            
            df = pd.DataFrame({'x': X_reduced[:, 0], 'y': X_reduced[:, 1], 'label': y_data})
            
            # 使用 seaborn 的 lmplot 进行绘图 
            sns.lmplot(data=df,  x='x', y='y', hue='label', fit_reg=False, 
                       height=8, aspect=1.5,  # 调整 aspect 使图像更宽 
                       palette="Set1", markers=['^', 'v', 's', 'o', '1'])
            
            plt.title(f"Perplexity:  {perplexity}, Iterations: {n_iter}") 
            img_name = f"{output_dir}/{img_name_prefix}_perp_{perplexity}_iter_{n_iter}.png"
            plt.savefig(img_name,  dpi=300)  # 增加图像分辨率
            plt.close() 
        except Exception as e: 
            print(f"An error occurred with perplexity {perplexity}: {e}")
 
# 假设 train 是已经加载的 DataFrame
X_pre_tsne = train.drop(['subject',  'Activity', 'ActivityName'], axis=1) 
y_pre_tsne = train['ActivityName'] 
 
# 调用函数执行 t-SNE
perform_tsne(X_data=X_pre_tsne, y_data=y_pre_tsne, perplexities=[2, 5, 10, 20, 50])
```
